```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("data.csv",sep=";")
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data["Target"]=encoder.fit_transform(data["Target"])
```

```python
plt.figure(figsize=(36,36))
cor_matrix1=data.corr(numeric_only=True)
sns.heatmap(cor_matrix1,annot=True,cmap=plt.cm.Reds)
plt.show()
```

```python
cor_target=abs(cor_matrix1["Target"])
cor_target
```

```python
relevant_features=cor_target[cor_target>0.38]
relevant_features
```

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeClassifier
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
model=DecisionTreeClassifier()
rfe=RFE(model,n_features_to_select=10)
fit=rfe.fit(x,y)
print("Num Features:",fit.n_features_)
print("Selected Features:",fit.support_)
print("Feature Ranking:",fit.ranking_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
Grade = np.array(data["Previous qualification (grade)"])
Marital_status = np.array(data["Marital status"])
plt.scatter(Grade, Marital_status, c=Grade, s=20) 
plt.title("Scatter Plot")
plt.xlabel("Grade")
plt.ylabel("Marital Status")
plt.show()
```
