```python
#4. Display the mean, median and mode for any two attributes of the dataset.

attributes=['Tuition fees up to date','Age at enrollment']
for attribute in attributes:
    # Calculate mean
    mean_value = df[attribute].mean()
    print(f"Mean {attribute}: {mean_value}")

    # Calculate median
    median_value = df[attribute].median()
    print(f"Median {attribute}: {median_value}")

    # Calculate mode
    mode_value = df[attribute].mode()
    
    print(f"Mode {attribute}: {mode_value.values[0]}")
    
```

```python
#20. Plot an attribute with different markers, different marker colors .

df = pd.DataFrame(data)

# Extract unique application modes
unique_modes = df['Application mode'].unique()

# Define markers and colors for each unique mode
markers = ['o', 's', '^', 'D', 'v']  # Different markers
colors = ['blue', 'green', 'red', 'purple', 'orange']  # Different marker colors

# Plotting with different markers and marker colors
plt.figure(figsize=(8, 6))

for i, mode in enumerate(unique_modes):
    mode_data = df[df['Application mode'] == mode]
    plt.scatter(mode_data.index, mode_data['Application order'], 
                marker=markers[i % len(markers)], color=colors[i % len(colors)], label=mode, s=100)

plt.title('Application mode vs Application order')
plt.xlabel('Index')
plt.ylabel('Application order')
plt.legend()

plt.grid(True)
plt.show()
```
